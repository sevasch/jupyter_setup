```python
x = np.linspace(-np.pi, np.pi, 1000)
y = np.sin(x) + np.random.normal(loc=0.0, scale=0.1, size=x.shape[0])

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, y, label="data")
ax.set_xlabel("x axis"), ax.set_ylabel("y axis")
ax.legend()
fig.tight_layout()
```
